```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(1,5,11)
y = x**2
```

```python
fig = plt.figure()
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**3, 'b--', label='x só')
ax.plot(x,x**4, 'g*--', label='y ao quadrado')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('titulo')
ax.legend(loc=0)
```
