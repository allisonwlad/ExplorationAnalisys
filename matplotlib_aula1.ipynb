<h1>Matplotlib<h1/>

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
# funcional 
plt.plot(x, y, color='r')
plt.xlabel('eixo X')
plt.ylabel('eixo Y')
plt.title('Grafico')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b*--')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axesX = fig.add_axes([0.1,0.1,0.8,0.8])
axesY = fig.add_axes([0.2,0.5,0.3,0.3])
axesX.plot(x, y)
axesX.set_xlabel('eixo x')
axesX.set_title('titulo')
axesY.plot(y,x)
```
